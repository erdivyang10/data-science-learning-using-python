```python
#Here we will create the facet_grid. 
#Facet grid forms a matrix of panels defined by row and column by dividing the variables.
#Due of panels, a single plot looks like multiple plots. 
#It is very helpful to analyze all combinations in two discrete variables.

#import pandas,matplotlib and seaborn library

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings

#read the csv file

movies = pd.read_csv('/home/flippercode/Documents/python-learning/include/movie-ratings.csv')
movies.head()

#increase the graph size

plt.rcParams['figure.figsize'] = 10,8

#remove the warning
warnings.filterwarnings('ignore')
```

```python
movies.columns
```

```python
#rename the fields of the csv file

movies.rename(columns={"Film":"Film","Genre":"Genre","Rotten Tomatoes Ratings %":"CriticRating",
                      "Audience Ratings %":"AudienceRating","Budget (million $)":"Budget",
                       "Year of release":"Year"},inplace=True)
```

```python
movies.columns
```

```python
vis2 = sns.lmplot(data=movies, x= 'CriticRating', y='AudienceRating',fit_reg=False, size=8, hue='Year')
```

```python
#Now we will create the facetgrid

g=sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
g= g.map(plt.scatter,'CriticRating','AudienceRating')

#Now we have created an amazing facet grid and displayed the results in the Year of each Gener
```

```python
plt.scatter(movies.CriticRating,movies.AudienceRating)
plt.show()
```

```python
#Now we will create a facet grid by using a Histogram"


g=sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
g= g.map(plt.hist,'CriticRating')
```

```python
#back to scatter
#Add a edge color and linewidth on the markers

g=sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50,linewidth=0.5, edgecolor='black')
g= g.map(plt.scatter,'CriticRating','AudienceRating',**kws)
```
