```python
#Here we will create coordinates and the diagonal line on each facet-grid


#import pandas,matplotlib and seaborn library

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings

#read the csv file

movies = pd.read_csv('/home/flippercode/Documents/python-learning/include/movie-ratings.csv')
movies.head()

#increase the graph size

plt.rcParams['figure.figsize'] = 10,8

#remove the warning
warnings.filterwarnings('ignore')
```

```python
movies.columns
```

```python
#rename the fields of the csv file

movies.rename(columns={"Film":"Film","Genre":"Genre","Rotten Tomatoes Ratings %":"CriticRating",
                      "Audience Ratings %":"AudienceRating","Budget (million $)":"Budget",
                       "Year of release":"Year"},inplace=True)
```

```python
movies.columns
```

```python
#Add a edge color and linewidth on the markers


g= sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50, linewidth=0.5, edgecolor='Black')
g= g.map(plt.scatter,'CriticRating','AudienceRating',**kws)

#start coordinates with 20 not with 0
g= g.set(xlim=(20,100),ylim=(20,100)) # graph values start with 20 not with 0

#Add the diagonal line so user can easily differentiate which one is Audience Rating and which one is Critic.
# we will use axes.flat function


for ax in g.axes.flat:
    ax.plot((20,100),(20,100),c='Gray', ls='--')

#add legend
g.add_legend()
    


```
