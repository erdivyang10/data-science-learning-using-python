```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')


#read the csv file
movies = pd.read_csv('/home/flippercode/Documents/python-learning/include/movie-ratings.csv')

#check the coloumns of the csv

movies.columns
```

```python
#rename the fields of the csv file

movies.rename(columns={"Film":"Film","Genre":"Genre","Rotten Tomatoes Ratings %":"CriticRating",
                      "Audience Ratings %":"AudienceRating","Budget (million $)":"Budget",
                       "Year of release":"Year"},inplace=True)
```

```python
movies.head()
```

```python
#creating KDE Kernel Density Graph

k = sns.kdeplot(movies.CriticRating,movies.AudienceRating)
```

```python
#Now we will create a 1x2 subplot

f, axes = plt.subplots(1,2,figsize=(12,6), sharex=True, sharey=True)
k1 = sns.kdeplot(movies.AudienceRating,movies.Budget, ax=axes[0] )
k2 = sns.kdeplot(movies.CriticRating,movies.AudienceRating, ax=axes[1], shade=True)
```

```python
#creating subplots with 2*2

f, axes = plt.subplots(2,2,figsize=(12,6), sharex=True, sharey=True)
k1 = sns.kdeplot(movies.AudienceRating,movies.Budget, ax=axes[0][1] )
k2 = sns.kdeplot(movies.CriticRating,movies.AudienceRating, ax=axes[1][0], shade=True)
k3 = sns.kdeplot(movies.AudienceRating,movies.Budget, ax=axes[0][0],cmap="Greens" )
k4 = sns.kdeplot(movies.CriticRating,movies.Budget, ax=axes[1][1], cmap='Reds')
```
