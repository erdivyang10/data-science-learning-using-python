```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')


#read the csv file
movies = pd.read_csv('/home/flippercode/Documents/python-learning/include/movie-ratings.csv')

#check the coloumns of the csv

movies.columns

```

```python
#rename the fields of the csv file

movies.rename(columns={"Film":"Film","Genre":"Genre","Rotten Tomatoes Ratings %":"CriticRating",
                      "Audience Ratings %":"AudienceRating","Budget (million $)":"Budget",
                       "Year of release":"Year"},inplace=True)
```

```python
#now check the newly name of Fields
movies.columns
```

```python
#Yeah that's the perfect field name
movies.head()
```

```python
#Now we will use the jointplot data

j = sns.jointplot(data = movies, x="CriticRating", y="AudienceRating")
```

```python
#Now display the diagram in the hexagonal by passing the kind paramter and set the hex
```

```python
k = sns.jointplot(data= movies, x="CriticRating",y="AudienceRating",kind="hex",height=6)
```

```python
#resid

l = sns.jointplot(data= movies, x="CriticRating",y="AudienceRating",kind="resid",)
```

```python
#reg
m = sns.jointplot(data= movies, x="CriticRating",y="AudienceRating",kind="reg",height=10)
```
