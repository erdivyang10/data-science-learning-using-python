```python
#import pandas,matplotlib and seaborn library

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings

#read the csv file

stats = pd.read_csv('/home/flippercode/Documents/python-learning/include/demographic.csv')
stats.head()

#increase the graph size

plt.rcParams['figure.figsize'] = 10,8

#remove the warning
warnings.filterwarnings('ignore')
```

```python
vis2 = sns.lmplot(data=stats, x= 'BirthRate', y='InternetUsers',fit_reg=False, size=8, hue='IncomeGroup')
```
