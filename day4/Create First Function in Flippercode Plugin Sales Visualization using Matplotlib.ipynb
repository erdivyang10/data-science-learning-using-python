```python
#Create a first function in Python:
#Here all plot values are repeated again and again, So we will optimize the code by us


def myplot(PluginList):
    Col = {'GoogleMapsPro':"Black",'UserAvatar':"Green",'StoreLocator':"Red",'Ecards':"Brown",'ReminderPro':"Blue",'CF7Extensions':"Yellow",'DeliveryArea':"Gray",'UpdatesPro':"Magenta",'ImageHover':"cyan",'SecurityQuestions':"skyblue",'BookingPopup':"black",'GDPR':"red"}
    Marker = {'GoogleMapsPro':"s",'UserAvatar':"p",'StoreLocator':"o",'Ecards':"v",'ReminderPro':".",'CF7Extensions':"d",'DeliveryArea':"D",'UpdatesPro':"P",'ImageHover':"*",'SecurityQuestions':"8",'BookingPopup':".",'GDPR':"h"}
    
    for PluginName in PluginList:
        plt.plot( Sales[PDict[PluginName]],  c=Col[PluginName], ls="--",  marker =Marker[PluginName], ms="10", label=PluginName)
    
    plt.rcParams['figure.figsize'] =10,10 # for size of the graph
    plt.legend( Plugin, loc='upper left' , bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,12)),Month)
    plt.show()
    
```

```python
def myplot(data,PluginList = Plugin):
    Col = {'GoogleMapsPro':"Black",'UserAvatar':"Green",'StoreLocator':"Red",'Ecards':"Brown",'ReminderPro':"Blue",'CF7Extensions':"Yellow",'DeliveryArea':"Gray",'UpdatesPro':"Magenta",'ImageHover':"cyan",'SecurityQuestions':"skyblue",'BookingPopup':"black",'GDPR':"red"}
    Marker = {'GoogleMapsPro':"s",'UserAvatar':"p",'StoreLocator':"o",'Ecards':"v",'ReminderPro':".",'CF7Extensions':"d",'DeliveryArea':"D",'UpdatesPro':"P",'ImageHover':"*",'SecurityQuestions':"8",'BookingPopup':".",'GDPR':"h"}
    
    for PluginName in PluginList:
        plt.plot( data[PDict[PluginName]],  c=Col[PluginName], ls="-",  marker =Marker[PluginName], ms="10", label=PluginName)
    
    plt.rcParams['figure.figsize'] =10,10 # for size of the graph
    plt.legend( loc='upper left' , bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,12)),Month)
    plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title("Sales on Each Month", fontdict=None, loc='center', pad=None)
print('Per Month Plugin Sales')
myplot(Sales)
plt.title("Sales Per Day in Each Month", fontdict=None, loc='center', pad=None)
myplot(SalePerDay)
```
