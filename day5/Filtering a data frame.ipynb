```python
#In this guide we will filter the content from the CSV file

import pandas as pd

#to read the CSV file

stats = pd.read_csv('/home/flippercode/Documents/python-learning/include/demographic.csv')
```

```python
stats
```

```python
#Filteration of Low Income countries
low_income = stats.IncomeGroup == 'Low income'
stats[low_income]
```

```python
#Filteration of High income countries
high_income = stats.IncomeGroup == 'High income'
stats[high_income]
```

```python
#Filteration of Countries which have less than 3 internet users

less_internet = stats.InternetUsers < 3
stats[less_internet]
```

```python
#Filteration of Countries which have birth rate greater than 40
birth_rate = stats.BirthRate > 45
stats[birth_rate]
```

```python
#filteration to get india

get_india= stats.CountryName == 'India'
stats[get_india]
```

```python
#if we need to filter the country where Birth rate is greater than 40 and internet user is less than 3

stats[(stats.BirthRate > 40) & (stats.InternetUsers < 3)]
```
